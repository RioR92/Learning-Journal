#Penerapan Looping untuk data frame

#siapkan list kosong, sebagai tempat penampung
#specify variable yang di butuhkan
#buat prosedur -> hasil
#hasilnya di simpan ke list kosong
#loop selesai, list sudah penuh -> convert menjadi kolom baru

```python
hasil = []
for i in range(0,len(nba)):
    selisih = nba.loc[i,'pts']-nba.loc[i,'opp_pts']
    hasil.append(selisih)

nba['selisih_loop']=hasil
nba
```

```python
# kolom baru hasil pertandingan
# def function
def hasil (selisih) :
    if selisih > 0:
        return 'menang'
    elif selisih < 0:
        return 'kalah'
    else :
        return 'seri'
```

```python
hasil_pertandingan = []

for i in range(0,len(nba)):
    result = hasil(nba.iloc[i,-2])
    hasil_pertandingan.append(result)

nba['hasil'] = hasil_pertandingan
```
